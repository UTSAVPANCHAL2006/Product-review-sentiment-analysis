```python
df_clean = df.drop(columns=['Id', 'ProductId', 'UserId', 'ProfileName', 'HelpfulnessNumerator', 'HelpfulnessDenominator', 'Time'])
```

```python

df_clean.head()
```

```python
df_clean['Summary'] = df_clean['Summary'].fillna('')
```

```python
df_clean.isnull().sum()
```

```python
df_clean['review'] = df_clean['Summary'] + ' ' + df_clean['Text']
```

```python
df_clean.drop(columns=['Summary', 'Text'], inplace=True)
```

```python
def sentiment(Score):
    if Score <= 2:
        return 'negative'
    elif Score == 3:
        return 'neutral'
    else:
        return 'positive'
    
df_clean['sentiment'] = df_clean['Score'].apply(sentiment)
```

```python
df_clean.head()
```

```python
import re
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r"http\S+|www\S+", "", text)
    text = re.sub(r"[^a-zA-Z\s]", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    words = [w for w in text.split() if w not in stop_words]
    return " ".join(words)

df_clean['cleaned_review'] = df_clean['review'].apply(clean_text)
```

```python
df_clean.drop(columns=['Score', 'review'], inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder  

le = LabelEncoder()
df_clean['label'] = le.fit_transform(df_clean['sentiment'])
df_clean.head()
```

```python
df_clean.drop(columns=['sentiment'], inplace=True)
```

```python
df_clean.to_csv('/Users/utsav/Desktop/Product-review-sentiment-analysis/data/clean-data.csv', index=False)
```
